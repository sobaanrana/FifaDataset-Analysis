## Task 4

Plotting graph to visualize number of players in clubs

```python
plt.figure(figsize = (20,10))

plt.style.use('fivethirtyeight')
ax = sns.countplot(x = 'Club', data = new_dataframe, palette = 'bone')
ax.set_xlabel(xlabel = 'Clubs', fontsize = 16)
ax.set_ylabel(ylabel = 'Count of Players', fontsize = 16)
ax.set_title(label = 'Comparsion of Positions and Players', fontsize = 20)
plt.yticks(size=20)
plt.xticks(size=10)
plt.show()
```

Plotting graph to visualize team position along with their counts

```python
plt.figure(figsize = (20,10))

plt.style.use('fivethirtyeight')
ax = sns.countplot(x = 'Position', data = new_dataframe, palette = 'bone')
ax.set_xlabel(xlabel = 'Position', fontsize = 16)
ax.set_ylabel(ylabel = 'Count of Players', fontsize = 16)
ax.set_title(label = 'Comparsion of Positions and Players', fontsize = 20)
plt.yticks(size=20)
plt.xticks(size=15)
plt.show()
```

## Task 5

Finding best player for LW position

```python
LW_list = new_dataframe["LW"].tolist()

playersList = []
for i in LW_list:
    value = 0
    if i ==0 :
        value = 0
        playersList.append(value)
        continue
    l = i.split("+")
    value = int(l[0])+int(l[1])
    playersList.append(value)
    
ind = playersList.index(max(playersList))

new_dataframe.iloc[ind]
```

The results show that L. Messi is the best LW position player.
